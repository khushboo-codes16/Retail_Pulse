```python
import mlflow
import mlflow.sklearn
import mlflow.pytorch
from prophet import Prophet
import joblib
import pandas as pd

# Set tracking URI (default is local)
mlflow.set_tracking_uri("http://localhost:5000")
mlflow.set_experiment("RetailPulse_Week1")
```

```python
print("Logging Prophet Model...")

# Load your trained model
prophet_model = joblib.load('../src/models/prophet_model.pkl')

with mlflow.start_run(run_name="Prophet_Forecast"):
    # Log parameters
    mlflow.log_param("model_type", "Prophet")
    mlflow.log_param("seasonality_mode", "multiplicative")
    mlflow.log_param("changepoint_prior_scale", 0.05)
    
    # Log metrics (you need to calculate these)
    # mlflow.log_metric("mape", prophet_mape)
    # mlflow.log_metric("rmse", prophet_rmse)
    
    # Log model
    mlflow.prophet.log_model(prophet_model, "prophet_model")
    
    # Log artifacts
    mlflow.log_artifact("../data/processed/forecast_comparison.png")
    
print("✅ Prophet model logged!")
```

```python
print("\nLogging LSTM Model...")

with mlflow.start_run(run_name="LSTM_Forecast"):
    mlflow.log_param("model_type", "LSTM")
    mlflow.log_param("hidden_size", 64)
    mlflow.log_param("num_layers", 2)
    mlflow.log_param("sequence_length", 30)
    mlflow.log_param("epochs", 100)
    
    # Log LSTM model (saved as .pth file)
    mlflow.log_artifact("../src/models/lstm_model.pth")
    mlflow.log_artifact("../src/models/lstm_scaler.pkl")
    
print("✅ LSTM model logged!")
```

```python
print("\nLogging Customer Segmentation Model...")

with mlflow.start_run(run_name="Customer_Segmentation"):
    mlflow.log_param("algorithm", "KMeans")
    mlflow.log_param("n_clusters", 6)
    mlflow.log_param("random_state", 42)
    
    # Log models
    mlflow.log_artifact("../src/models/kmeans_segmenter.pkl")
    mlflow.log_artifact("../src/models/rfm_scaler.pkl")
    
    # Log segmentation results
    segments = pd.read_csv('../data/processed/customer_segments.csv')
    mlflow.log_metric("n_customers", len(segments))
    mlflow.log_metric("n_segments", segments['Segment'].nunique())
    
print("✅ Segmentation model logged!")
```

```python
print("\nLogging Churn Prediction Model...")

with mlflow.start_run(run_name="Churn_Prediction"):
    mlflow.log_param("algorithm", "XGBoost")
    mlflow.log_param("n_estimators", 200)
    mlflow.log_param("max_depth", 6)
    mlflow.log_param("learning_rate", 0.05)
    
    # Log model
    mlflow.log_artifact("../src/models/churn_xgboost.pkl")
    
print("✅ Churn model logged!")
```

```python
print("\n" + "="*50)
print("All MLflow runs completed!")
print("View at: http://localhost:5000")
print("="*50)
```
